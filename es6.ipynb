```python
import numpy as np
import matplotlib.pyplot as plt

N, E, stdev_E = np.loadtxt("es6/output.chi.0", usecols=(0,1,2), delimiter=',', unpack='true')

plt.figure(figsize=(14,6))
plt.errorbar(N,E,yerr=stdev_E, marker='', ecolor='orange')
plt.xlabel('N')
plt.ylabel('E')
plt.grid(True)
plt.show()
```

```python
N, E = np.loadtxt("es6/m_values.0", usecols=(0,1), delimiter=',', unpack='true')

plt.figure(figsize=(14,6))
plt.plot(N,E)
plt.xlabel('N')
plt.ylabel('u')
plt.grid(True)
plt.show()
```
