## Esercizio 4.1

Mostro che il mio sistema Ã¨ stato termodinamicamente equilibrato, come si osserva dai valori grafici di temperatura e energia nel progredire della simulazione. 

```python
import numpy as np
import matplotlib.pyplot as plt

Temp = np.loadtxt("es04/punto1/output_temp.dat", usecols=(0), unpack='true')
Etot = np.loadtxt("es04/punto1/output_etot.dat", usecols=(0), unpack='true')
Epot = np.loadtxt("es04/punto1/output_epot.dat", usecols=(0), unpack='true')
Ekin = np.loadtxt("es04/punto1/output_ekin.dat", usecols=(0), unpack='true')

a = np.arange(1000)

plt.figure(figsize=(14,8))
plt.plot(a,Temp, label='Temperature (LJ)')
plt.plot(a,Etot, label='Total energy (LJ)')
plt.plot(a,Epot, label='Potential energy (LJ)')
plt.plot(a,Ekin, label='Kinetic Energy (LJ)')
plt.xlabel('step')
plt.legend(frameon = True)
plt.grid(True)
plt.show()
```

## Esercizio 4.3

Mostro l'andamento delle grandezze medie per particella in funzione del numero di blocchi, in unitÃ  SI. Eseguo le 3 simulazioni in sequenza, mostrando anche come il programma riequilibri correttamente le grandezze partendo da ogni condizione.

### Configurazione 1: solido


```python
Nblocks, Epot, var_Epot = np.loadtxt("es04/3_solid/ave_epot.out", usecols=(0,1,2) , delimiter=',', unpack='true')

print ('                                              Solid state')
fig = plt.figure(figsize=(14,15))

ax1 = fig.add_subplot(221)
ax1.errorbar(Nblocks,Epot,yerr=var_Epot**0.5, marker='', ecolor='red')
ax1.set_xlabel('Nblocks')
ax1.set_ylabel('Epot/N (J)')
ax1.grid(True)

Nblocks, Ekin, var_Ekin = np.loadtxt("es04/3_solid/ave_ekin.out", usecols=(0,1,2) , delimiter=',', unpack='true')

ax2 = fig.add_subplot(222)
ax2.errorbar(Nblocks,Ekin,yerr=var_Ekin**0.5, marker='', ecolor='red')
ax2.set_xlabel('Nblocks')
ax2.set_ylabel('Ekin/N (J)')
ax2.grid(True)

Nblocks, Etot, var_Etot = np.loadtxt("es04/3_solid/ave_etot.out", usecols=(0,1,2) , delimiter=',', unpack='true')

ax3 = fig.add_subplot(223)
ax3.errorbar(Nblocks,Etot,yerr=var_Etot**0.5, marker='', ecolor='red')
ax3.set_xlabel('Nblocks')
ax3.set_ylabel('Etot/N (J)')
ax3.grid(True)

Nblocks, Temp, var_Temp = np.loadtxt("es04/3_solid/ave_temp.out", usecols=(0,1,2) , delimiter=',', unpack='true')

ax4 = fig.add_subplot(224)
ax4.errorbar(Nblocks,Temp,yerr=var_Temp**0.5, marker='', ecolor='red')
ax4.set_xlabel('Nblocks')
ax4.set_ylabel('Temp (K)')
ax4.grid(True)
```
