# Esercitazione 4


Il semplice comando _make_ esegue le seguenti operazioni:

- elimina i file delle posizioni e dei parametri di output creati al test precedente
- salva in _config.0_ la _config.final_ del test precedente. In runtime viene chiesto all'utente se ripartire dallo step t-dt e se rinormalizzare tali posizioni per adattare alla temperatura voluta.
- esegue il programma

Inoltre, con _make solid_ e _make liquid_ vengono caricate in _input.dat_ le configurazioni solide o liquide fornite.


## Esercizio 4.1

Mostro che il mio sistema Ã¨ stato termodinamicamente equilibrato, come si osserva dai valori grafici di temperatura e energia nel progredire della simulazione. 

```python
import numpy as np
import matplotlib.pyplot as plt

Temp = np.loadtxt("es4/punto1/output_temp.dat", usecols=(0), unpack='true')
Etot = np.loadtxt("es4/punto1/output_etot.dat", usecols=(0), unpack='true')
Epot = np.loadtxt("es4/punto1/output_epot.dat", usecols=(0), unpack='true')
Ekin = np.loadtxt("es4/punto1/output_ekin.dat", usecols=(0), unpack='true')

a = np.arange(1000)

plt.figure(figsize=(14,8))
plt.plot(a,Temp, label='Temperature (LJ)')
plt.plot(a,Etot, label='Total energy (LJ)')
plt.plot(a,Epot, label='Potential energy (LJ)')
plt.plot(a,Ekin, label='Kinetic Energy (LJ)')
plt.xlabel('step')
plt.legend(frameon = True)
plt.grid(True)
plt.show()
```

### Configurazione 1: solido


```python
Nblocks, Epot, var_Epot = np.loadtxt("es4/3_1/ave_epot.out", usecols=(0,1,2) , delimiter=',', unpack='true')

plt.figure(figsize=(14,6))
plt.errorbar(Nblocks,Epot,yerr=var_Epot**0.5, marker='', ecolor='red')
plt.xlabel('Nblocks')
plt.ylabel('Epot/N (J)')
plt.grid(True)
plt.show()
```
