Implemento una classe _black_scholes_ in cui inserisco tutte le funzioni necessarie per il calcolo delle opzioni call e put, che calcolo direttamente e in maniera discreta (impiegando 100 passi). 

Di seguito i risultati delle simulazioni, che confronto con la soluzione analitica.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Call option

```python
Nblocks, Cd, var_Cd = np.loadtxt("es03/Cd.out", usecols=(0,1,2) , delimiter=',', unpack='true')
Nblocks, Cc, var_Cc = np.loadtxt("es03/Cc.out", usecols=(0,1,2) , delimiter=',', unpack='true')

x = np.arange(101)
y = [14.975790778311286]*101

plt.figure(figsize=(16,10))
plt.errorbar(Nblocks,Cd,yerr=var_Cd**0.5, marker='', ecolor='red', label='direct')
plt.errorbar(Nblocks,Cc,yerr=var_Cc**0.5, marker='', ecolor='green', label='discretized')
plt.plot(x, y, color = 'purple', label='analyitic')
plt.xlabel('N')
plt.ylabel('C(T)')
plt.legend()
plt.grid(True)
plt.show()
```

### Put option

```python
Nblocks, Pd, var_Pd = np.loadtxt("es03/Pd.out", usecols=(0,1,2) , delimiter=',', unpack='true')
Nblocks, Pc, var_Pc = np.loadtxt("es03/Pc.out", usecols=(0,1,2) , delimiter=',', unpack='true')

y = [5.4595325819072364]*101

plt.figure(figsize=(16,10))
plt.errorbar(Nblocks,Pd,yerr=var_Pd**0.5, marker='', ecolor='red', label='direct')
plt.errorbar(Nblocks,Pc,yerr=var_Pc**0.5, marker='', ecolor='green', label='discretized')
plt.plot(x, y,color = 'purple', label='analyitic')
plt.xlabel('N')
plt.ylabel('P(T)')
plt.legend()
plt.grid(True)
plt.show()
```

I valori finali ottenuti appaiono in generale compatibili tra loro e con il calcolo analitico, seppur il valore finale della Call option calcolata direttamente si discosti più di $1 \sigma$ dal valore analitico.
