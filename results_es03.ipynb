## Esercizio 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Call option, discrete case at t=T


```python
Nblocks, C, var_C = np.loadtxt("es03/Cd.out", usecols=(0,1,2) , delimiter=',', unpack='true')

plt.figure(figsize=(14,6))
plt.errorbar(Nblocks,C,yerr=var_C**0.5, marker='', ecolor='red')
plt.xlabel('N')
plt.ylabel('C(T)')
plt.grid(True)
plt.show()
```
