Per trovare il passo $\delta$ tale da avere accettanza 50%, eseguo innanzitutto il programma variando tale parametro, separatamente per le due funzioni e per i due metodi di movimento T (con prob. uniforme o gaussiana). Individuo quindi i valori corretti, che fornisco tramite _input.dat_.

Scelgo come posizione di partenza un punto in cui la probabilità $|\psi|^2$ non è eccessivamente bassa, così che l'algoritmo non impieghi troppi step a tornare nella regione in cui la funzione d'onda ha valori più alti. Se parto molto lontano dall'origine, l'algoritmo torna comunque in tale regione impiegando tempi lunghi, a meno che si arrivi ai limiti di funzionalità del calcolatore.

Ho fissato a 100 il numero di step per l'algoritmo di Metropolis; avendo un'accettanza del 50% vuol dire che ogni valore $<r>$  viene ottenuto con circa 50 passi. Non osservo miglioramenti nei risultati aumentando il numero di step, pertanto mi fermo a 100 per non allungare eccessivamente i tempi di computazione.

###  $\psi_{100}$, con T uniforme
($\delta= 1.2$)

```python
import numpy as np
import matplotlib.pyplot as plt

N, r, stdev_U1 = np.loadtxt("es05/U1.out", usecols=(0,1,2), delimiter=',', unpack='true')

plt.figure(figsize=(14,6))
plt.errorbar(N,r,yerr=stdev_U1, marker='', ecolor='orange')
plt.xlabel('N')
plt.ylabel('<r> - 1.5')
plt.grid(True)
plt.show()
```

Osservo che i valori assoluti di $<r>$ risultano più vicini a quelli attesi nel caso gaussiano, sia per $\psi_{100}$ che per $\psi_{210}$. Confrontando (di seguito) le deviazioni standard, non noto una significativa differenza tra T uniforme e gaussiano.
