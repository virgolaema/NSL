Per trovare il passo delta tale da avere accettanza 50%, eseguo il programma variando il passo delta, separatamente per le due funzioni e per i due metodi di movimento con T (prob. uniforme e gaussiana).

Scelgo come posizione di partenza un punto in cui la probabilità non è eccessivamente bassa, così che l'algoritmo non impieghi troppi step a tornare nella regione in cui la funzione d'onda ha valori più alti. Se parto lontano dall'origine, l'algoritmo torna comunque in tale regione ma ci mette tanti steps. Se parto eccessivamente lontano, c'è il rischio che la p arrivi al limite del calcolatore e che l'algoritmo smetta di funzionare opportunamente.

Ho messo come step per Metropolis 100. Siccome il tasso di accettazione è stato fissato tramite delta al 50%, vuol dire che ogni valore di R utilizzato viene da circa 50 passi.

```python
import numpy as np
import matplotlib.pyplot as plt


#Caso 100, con T uniforme
N, r, stdev_r = np.loadtxt("es5/U1.out", usecols=(0,1,2), delimiter=',', unpack='true')

plt.figure(figsize=(14,6))
plt.errorbar(N,r,yerr=stdev_r, marker='', ecolor='orange')
plt.xlabel('N')
plt.ylabel('<r> - 1.5')
plt.grid(True)
plt.show()
```
