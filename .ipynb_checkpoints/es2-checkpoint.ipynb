## Esercizio 2.1

### Punto 1


```python
import numpy as np
import matplotlib.pyplot as plt

Nblocks,I,var_I,stdev_I= np.loadtxt("es2_1/es1.txt", usecols=(0,1,2,3) , delimiter=',', unpack='true')

plt.figure(figsize=(14,6))
plt.errorbar(Nblocks,I-1,yerr=stdev_I, marker='s', ecolor='red')
plt.xlabel('N')
plt.ylabel('<I>-1')
plt.grid(True)
plt.show()
```

```python
from scipy.stats import expon
from scipy import optimize

def test_func(x, a):
    return a/np.sqrt(x)

params, params_covariance = optimize.curve_fit(test_func, Nblocks, stdev_I) #, p0=[2, 2]
plt.figure(figsize=(14,6))

plt.plot(Nblocks,stdev_I, label='Data')
plt.xlabel('N')
plt.ylabel('$\sigma_I$')
plt.grid(True)

plt.plot(Nblocks, test_func(Nblocks, params[0]),label='Fitted function')
plt.legend(loc='best')

plt.show()

print ('Fitto con una curva del tipo a/sqrt(N), che risulta in buon accordo con i dati.')
```

## Esercizio 2

### Punto 1

```python
i,f,stdev = np.loadtxt("es2_2/es1.txt", usecols=(0,1,2) , delimiter=',', unpack='true')

def test_func(x, a):
    return a * np.sqrt(x)

params = optimize.curve_fit(test_func, i, f, p0=[1])

plt.figure(figsize=(14,10))
plt.errorbar(i,f,yerr=stdev, ecolor='red', label='Data')
plt.plot(i, test_func(i, params[0]),label='Fitted function', color='orange')
plt.legend(loc='best')
plt.xlabel('#steps')
plt.ylabel('f')
plt.grid(True)
plt.show()
```

Il fit di una funzione del tipo $f = k \sqrt{N}$ presenta un ottimo accordo con i dati.
