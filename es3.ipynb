```python
import numpy as np
import matplotlib.pyplot as plt

#Call option, discrete case at t=T
Nblocks, C, stdev_C = np.loadtxt("es3/Cd.txt", usecols=(0,1,2) , delimiter=',', unpack='true')

plt.figure(figsize=(14,6))
plt.errorbar(Nblocks,C,yerr=stdev_C, marker='s', ecolor='red')
plt.xlabel('N')
plt.ylabel('C(T)')
plt.grid(True)
plt.show()
```
