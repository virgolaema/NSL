```python
import numpy as np
import pandas
import matplotlib.pyplot as plt

es21 = pandas.read_csv('es2_1/es2_1.csv')
es21.head(10) #showing the first 10 values
```

```python
plt.errorbar(es21['Nblocks'],es21['mean_I'],yerr=es21['stdev_I'])
plt.xlabel('N')
plt.ylabel('I')
plt.grid(True)
plt.show()
```

```python
plt.plot(es21['Nblocks'],es21['stdev_I'])
plt.xlabel('N')
plt.ylabel('stdev_I')
plt.grid(True)

t = np.arange(2.0, 100.0, 0.2)
s = 1.5*1e-2/np.sqrt(t)

plt.plot(t,s)

plt.show()
```
