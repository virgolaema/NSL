### Punto 1


```python
import numpy as np
import matplotlib.pyplot as plt

Nblocks,I,var_I,stdev_I= np.loadtxt("es2_1/es1.txt", usecols=(0,1,2,3) , delimiter=',', unpack='true')

plt.figure(figsize=(14,6))
plt.errorbar(Nblocks,I-1,yerr=stdev_I, marker='s', ecolor='red')
plt.xlabel('N')
plt.ylabel('<I>-1')
plt.grid(True)
plt.show()
```

```python
from scipy.stats import expon
from scipy import optimize

def test_func(x, a):
    return a/np.sqrt(x)

params, params_covariance = optimize.curve_fit(test_func, Nblocks, stdev_I) #, p0=[2, 2]
plt.figure(figsize=(14,6))

plt.plot(Nblocks,stdev_I, label='Data')
plt.xlabel('N')
plt.ylabel('$\sigma_I$')
plt.grid(True)

plt.plot(Nblocks, test_func(Nblocks, params[0]),label='Fitted function')
plt.legend(loc='best')

plt.show()

print ('Fitto con una curva del tipo a/sqrt(N), che risulta in buon accordo con i dati.')
```
