## Punto 1

Valuto come varia la stima della media (con relativa varianza) al variare del numero di valori (aggiunti a blocchi) per calcolare la media. 



```python
import numpy as np
import pandas
import matplotlib.pyplot as plt

es11 = pandas.read_csv('es1_1/es1_1.csv')
es11.head(10) #showing the first 10 values
```

```python
plt.errorbar(es11['throws'],es11['mean_r']-0.5,yerr=es11['stdev_r'])
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

### Punto 2

Valuto ora come varia la varianza, e la varianza sulla stima della varianza.

```python
plt.errorbar(es11['throws'], es11['mean_var']-1/12., yerr=es11['stdev_var'])
plt.xlabel('#throws')
plt.ylabel('<(r-0.5)^2>-1/12')
plt.grid(True)
plt.show()
```

```python
fig, axs = plt.subplots(1, 2)

# We can set the number of bins with the `bins` kwarg
axs[0].hist(es12['standard'], bins=20)
axs[1].hist(es12['exponential'], bins=20)
```

```python
from scipy.stats import norm
import matplotlib.mlab as mlab

plt.hist(es12['lorentz'], bins=20, range=(-15,15))
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(es12['exponential'], bins=20)

```

```python
n = 10000
appo = np.arange(n)
for i in range(n):
    appo[i] = es12['lorentz'][i]

print (appo)
(mu, sigma) = norm.fit(appo)

# add a 'best fit' line
y = norm(mu,sigma)
x = np.arange(0,10,50)
l = plt.plot(x, y.pdf(x), '--r',linewidth=0.2)
#plt.hist(es12['exponential'], bins=50)

plt.grid(True)
plt.show()
```
