```python
#N=10
standard, expo, lorentz = np.loadtxt('es1_2/es2_10.txt', usecols=(0,1,2), delimiter=',', unpack='true')

fig, (a1,a2,a3) = plt.subplots(nrows=1, ncols=3, figsize=(16,5))
a1.hist(standard, bins=20)
a2.hist(expo, bins=20)
a3.hist(lorentz, bins=20, range=(-15,15))

print ('Caso N=10')

plt.show()
```

```python
#N=100
from scipy.stats import norm,cauchy

standard, expo, lorentz = np.loadtxt('es1_2/es2_100.txt', usecols=(0,1,2), delimiter=',', unpack='true')
n_bins=30

fig, (a1,a2,a3) = plt.subplots(nrows=1, ncols=3, figsize=(16,5))

_, nbins, _ = a1.hist(standard, n_bins, density=1, alpha=0.5)
mu, sigma = norm.fit(standard)
best_fit_line = norm.pdf(nbins, mu, sigma)
a1.plot(nbins,best_fit_line)

_, nbins, _ = a2.hist(expo, n_bins, density=1, alpha=0.5)
mu, sigma = norm.fit(expo)
best_fit_line = norm.pdf(nbins, mu, sigma)
a2.plot(nbins,best_fit_line)

_, nbins, _ = a3.hist(lorentz, n_bins, density=1, alpha=0.5, range=(-15,15))
mean, var= cauchy.fit(lorentz)
best_fit_line = cauchy.pdf(nbins, mean, var)
a3.plot(nbins,best_fit_line)

print ('Caso N=100')

plt.show()

print ('Ho osservato come per la terza distribuzione non sia possibile un fit gaussiano, \
       mentre il fit con una distribuzione di Cauchy appare buono')
```
